```python
word_bank = ["jazz", "hajj", "fuzzy", "glyph", "ivory", "unzip"]
word_select = 4
secret_word = []
for i in word_bank[word_select]:
    secret_word.append(i)
lives = 10
visual = []
for i in range(len(word_bank[word_select])):
    visual.append("_")
print(word_bank[word_select])
print(secret_word)
```

```python
visual = []
previous_guesses = []
for i in range(len(word_bank[word_select])):
    visual.append("_")
runcount = 0
win_flag = 0
lives = 3 #duplicate of lives initialisation to reset easily in test phase
print(" ".join(visual))
while lives > 0 and runcount < 50 and win_flag != 1:
    correct_guess_flag = False
    runcount += 1
    guess = input("Please enter a letter").lower()
    if guess not in previous_guesses:
        if len(guess) > 1 or type(guess) != str:
            validation_check = False
            print("Invalid letter input. Please enter a single letter of the alphabet. \n")
        else:
            validation_check = True
            for i in range(len(secret_word)):
                if guess == secret_word[i]:
                    correct_guess_pos = i
                    visual[correct_guess_pos] = guess
                    correct_guess_flag = True
            print(" ".join(visual))
            previous_guesses.append(guess)
            if correct_guess_flag == False:
                lives -= 1
                print("Incorrect. Minus one life. Remaining lives =", lives)
            if "_" not in visual:
                print("Congratulations! You won with {} lives remaining. Your word was {}".format(lives,word_bank[word_select]))
                win_flag = 1
    else:
        print("Letter has already been guessed")
```
