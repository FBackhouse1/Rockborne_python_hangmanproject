## Session Statistics

```python
# Session tracking global statistics
# placed here to persist across multiple rounds of the game
# and initialised only at the start of each session
total_wins = 0
games_played = 0
```

## Game Configuration and Word Bank/Selection Logic

```python
# import package for random number generation
import random

# dictionary containing word list for selection by category
word_bank_full = {
    "Animals": ["ant", "baboon", "badger", "bat", "bear", "beaver", "camel", "cat", "clam", "cobra", "cougar", "coyote", "crow", "deer", "dog", "donkey", "duck", "eagle", "ferret", "fox", "frog", "goat", "goose", "hawk", "lion", "lizard", "llama", "mole", "monkey", "moose", "mouse", "mule", "newt", "otter", "owl", "panda", "parrot", "pigeon", "python", "rabbit", "ram", "rat", "raven", "rhino", "salmon", "seal", "shark", "sheep", "skunk", "sloth", "snake", "spider", "stork", "swan", "tiger", "toad", "trout", "turkey", "turtle", "weasel", "whale", "wolf", "wombat", "zebra"],
    "Geography": ["africa", "albania", "algeria", "andes", "arctic", "asia", "austria", "belgium", "brazil", "canada", "china", "egypt", "europe", "france", "germany", "india", "italy", "japan", "london", "mexico", "norway", "paris", "russia", "spain", "tokyo"],
    "Coding": ["algorithm", "binary", "boolean", "compiler", "database", "function", "integer", "iterator", "keyword", "variable", "syntax", "string", "loop", "python", "github"]
}

def get_user_category():
    """
    Description: Displays a menu and captures the user's category choice.
    Parameters: None
    Outputs: 
        - str: The name of the selected category (e.g., 'Animals').
    """
    categories = list(word_bank_full.keys()) # capable of taking additional categories if added to full word bank
    print("--- SELECT A CATEGORY ---")
    for i, name in enumerate(categories, 1):
        print(f"{i}. {name}") 
    
    while True: # will continue asking for inputs until valid category number entered
        choice = input("\nEnter number (1-{}): ".format(len(categories)))
        if choice.isdigit() and 1 <= int(choice) <= len(categories): # input validation checks
            return categories[int(choice) - 1]
        print(f"Invalid input. Please enter a number between 1 and {len(categories)}.")

def initialize_word(category_name):
    """
    Description: Selects a random word and creates the initial lists for the game.
    Parameters:
        - category_name (str): The key to look up in word_bank_full.
    Outputs:
        - secret_word_list (list): The hidden word to guess in list format
        - visual_list (list): The visual display of the word to be guessed in list format
        - chosen_word (str): Hidden word in string format
    """
    word_list = word_bank_full[category_name]
    word_select_index = random.randint(0, len(word_list)-1) # word is randomly selected from user chosen category
    chosen_word = word_list[word_select_index]
    
    secret_word_list = [i for i in chosen_word] # word to be guessed converted to list format for comparison logic
    visual_list = ["_" for _ in range(len(chosen_word))] # separate list created for visual display of word during guessing
    
    return secret_word_list, visual_list, chosen_word
```

## Main Gameplay Function

```python
def play_hangman():
    """
    Description: The core game loop that manages turns, input validation, and score updates.
    Parameters: None (Uses global variables for session tracking).
    Outputs: None (Prints game status and updates global counters).
    """
    # Access global session statistics to update based on game outcome
    global total_wins, games_played
    
    # 1. Setup word to be guessed - call previous functions
    user_cat = get_user_category()
    secret_word, visual, full_word_string = initialize_word(user_cat)
    
    # 2. Local game variables, re-initialised every play
    lives = 5 
    previous_guesses = []
    runcount = 0
    win_flag = 0
    
    print(" ".join(visual))
    
    # 3. Main gameplay loop
    while lives > 0 and runcount < 50 and win_flag != 1:
        
        correct_guess_flag = False # reset correct guess flag
        runcount += 1 # track runcount to prevent potential infinite loop error
        
        guess = input("Please enter a letter: ").lower() # user guess for hangman letter

        # Guess validation checks
        if guess not in previous_guesses: # user can't repeat guesses
            if len(guess) > 1 or not isinstance(guess, str): # user can only guess single letters
                print("Invalid letter input. Please enter a single letter of the alphabet. \n")
           
            # If guess valid
            else: 
                for i in range(len(secret_word)):
                    if guess == secret_word[i]:
                        visual[i] = guess # visual display reflects correctly guessed letters
                        correct_guess_flag = True # if guessed letter appears in word a single time, flag set to true
                
                print(" ".join(visual)) # print updated visual display
                previous_guesses.append(guess)

                # If guessed letter does not appear in word
                if not correct_guess_flag:
                    lives -= 1
                    if lives == 0:
                        print("Incorrect. No remaining lives. You lose. Your word was", full_word_string)
                    else: 
                        print("Incorrect. Minus one life. Remaining lives =", lives)
                
                if "_" not in visual: # checks if all letters in word have been guessed
                    print("Congratulations! You won with {} lives remaining. Your word was {}".format(lives, full_word_string))
                    win_flag = 1 # win flag updated to exit loop
                    total_wins += 1 # global variable total_wins updated for session statistics tracking
        else:
            print("Letter has already been guessed")
    
    # 4. Final session update
    games_played += 1 # global variable update for session tracking
    win_rate = 100 * total_wins / games_played
    print("\nYour current stats for the session are: {} wins out of {} games played - {}% win rate".format(total_wins, games_played, win_rate))
```

## Play the Game

```python
# Calls main gameplay function - starts the game
play_hangman()
```
