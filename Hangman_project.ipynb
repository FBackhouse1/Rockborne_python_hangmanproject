```python
import random

word_bank_full = {
    "Animals": ["ant", "baboon", "badger", "bat", "bear", "beaver", "camel", "cat", "clam", "cobra", "cougar", "coyote", "crow", "deer", "dog", "donkey", "duck", "eagle", "ferret", "fox", "frog", "goat", "goose", "hawk", "lion", "lizard", "llama", "mole", "monkey", "moose", "mouse", "mule", "newt", "otter", "owl", "panda", "parrot", "pigeon", "python", "rabbit", "ram", "rat", "raven", "rhino", "salmon", "seal", "shark", "sheep", "skunk", "sloth", "snake", "spider", "stork", "swan", "tiger", "toad", "trout", "turkey", "turtle", "weasel", "whale", "wolf", "wombat", "zebra"],
    "Geography": ["africa", "albania", "algeria", "andes", "arctic", "asia", "austria", "belgium", "brazil", "canada", "china", "egypt", "europe", "france", "germany", "india", "italy", "japan", "london", "mexico", "norway", "paris", "russia", "spain", "tokyo"],
    "Coding": ["algorithm", "binary", "boolean", "compiler", "database", "function", "integer", "iterator", "keyword", "variable", "syntax", "string", "loop", "python", "github"]
}
categories = list(word_bank_full.keys())

print("--- SELECT A CATEGORY ---")
for i, name in enumerate(categories, 1):
    print(f"{i}. {name}")

while True:
    choice = input("\nEnter number (1-{}): ".format(len(categories)))
    
    # Check if input is a digit and within range
    if choice.isdigit() and 1 <= int(choice) <= len(categories):
        choice_index = int(choice) - 1
        break
    else:
        print(f"Invalid input. Please enter a number between 1 and {len(categories)}.")

user_cat = categories[choice_index]
word_bank = word_bank_full[user_cat]
word_select = random.randint(0, len(word_bank)-1)

secret_word = []
for i in word_bank[word_select]:
    secret_word.append(i)
lives = 10
visual = []
for i in range(len(word_bank[word_select])):
    visual.append("_")
print(word_bank[word_select])
print(secret_word)
```

```python
visual = []
previous_guesses = []
for i in range(len(word_bank[word_select])):
    visual.append("_")
runcount = 0
win_flag = 0
lives = 3 #duplicate of lives initialisation to reset easily in test phase
print(" ".join(visual))
while lives > 0 and runcount < 50 and win_flag != 1:
    correct_guess_flag = False
    runcount += 1
    guess = input("Please enter a letter").lower()
    if guess not in previous_guesses:
        if len(guess) > 1 or type(guess) != str:
            validation_check = False
            print("Invalid letter input. Please enter a single letter of the alphabet. \n")
        else:
            validation_check = True
            for i in range(len(secret_word)):
                if guess == secret_word[i]:
                    correct_guess_pos = i
                    visual[correct_guess_pos] = guess
                    correct_guess_flag = True
            print(" ".join(visual))
            previous_guesses.append(guess)
            if correct_guess_flag == False:
                lives -= 1
                print("Incorrect. Minus one life. Remaining lives =", lives)
            if "_" not in visual:
                print("Congratulations! You won with {} lives remaining. Your word was {}".format(lives,word_bank[word_select]))
                win_flag = 1
    else:
        print("Letter has already been guessed")
```
